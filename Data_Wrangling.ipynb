```python
import pandas as pd
import numpy as np
```

```python
df_pesquisa = pd.read_csv("pesquisa de pessoas.csv")
```

# Tratamento dos dados

```python
df_pesquisa.head(5)
```

```python
# Dropa colunas desnecessárias
df_pesquisa = df_pesquisa.drop(['Unnamed: 0'], axis=1) 
```

```python
# Renomeia colunas 

df_pesquisa = df_pesquisa.rename(columns={
                'Endereerço': 'Endereço',
                'phone': 'Telefone',
                'anual_income': 'Renda anual',
                'AAge': 'Idade',
                'n_filhos': 'Qntd de filhos',
                'Food_restrictions': 'Restrição alimentar',
                'diariamente_sun blocker': 'Utiliza protetor solar'
})
```

Expressão regular: O **\W** equivale a [^a-zA-Zà-úÀ-Ú0-9_].

Corresponde a todos os caracteres que não são letras maiusculas ou minusculas.

```python
# Remoção de caracteres especiais da coluna de nome
# Expressão regular: O **\W** equivale a [^a-zA-Zà-úÀ-Ú0-9_]. Corresponde a todos os caracteres que não são letras maiusculas ou minusculas substituido por ''

df_pesquisa['Nome'] = df_pesquisa['Nome'].str.replace('[^a-zA-Zà-úÀ-Ú0 ]','')
```

```python
# Organização da coluna de telefone
# Expressão regular que substitui todos os caracteres que não são numeros de 0-9 sinal de + e - por *

df_pesquisa['Telefone'] = df_pesquisa['Telefone'].str.replace('[^(0-9) +-]','*')
```

```python
# Remoção de telefones com menos de 10 numeros setado como * na organização da coluna
df_pesquisa['Telefone'].str.contains('\*')
```

```python
df_pesquisa.loc[(df_pesquisa['Telefone'].notna()) & (df_pesquisa['Telefone'].str.contains('\*')), 'Telefone'] = np.nan # Transforma os valores nan em False e transforma em Nan apenas os valores que possuem * se não excluir Nan ocorre o erro Cannot mask with non-boolean array containing NA / NaN values

# df_pesquisa.loc[df_pesquisa.str.contains('\*', na=False), 'Telefone'] = np.nan
```

```python
df_pesquisa.head(6)
```

```python
# Padronização da coluna Estado Civil

df_pesquisa['Estado Civil'].value_counts()
```

```python
df_pesquisa['Estado Civil'] = df_pesquisa['Estado Civil'].replace({
            'D': 'Divorciado',
            'S': 'Solteiro',
            'V': 'Viúvo',
            'C': 'Casado'
    
})
```

#### Cenário 1:

A empresa Carri Construtora contratou a empresa que você trabalha para encontrar possíveis compradores para o seus novos empreendimentos.

A empresa quer entender as necessidades dos clientes, e quer informações com:

*   Quais cliente devemos abordar;
*   Qual empreendimento nós devemos mostrá-los;
*   Esse cliente está em busca em investir em um imóvel ou comprar para moradia?


 Se renda anual > 50 mil = pode participar do Minha casa minha vida


```python
# Quais cliente devemos abordar: Movel quitado e financiado=investimento; Movel alugado, de favor=Moradia
# Seleção apenas de clientes com Renda anual > 50
```

```python
df_pesquisa.columns
```

```python
# Remover os dados que não importam para o negócio
df_construct = df_pesquisa.drop(['Restrição alimentar','streaming','Utiliza protetor solar','Compra e Lê Livros Todos os Anos','Quantidade de Livros Comprados',
       'Autor Favorito'], axis=1)
df_construct.head(5)
```

```python
# Seleção de clientes apenas com renda anual maior que 50 mil
df_construct = df_construct.loc[df_construct['Renda anual'] > 50000]
```

```python
df_construct.shape
```

```python
# Abordar clientes com moveis quitados e financiados para investimento
df_construct['Investimento'] = np.where((df_construct['Moradia'] == 'Quitada') | (df_construct['Moradia'] == 'Financiamento'), True, False)
```

```python
df_construct.head(5)
```

```python
# Abordar clientes com Movel alugado ou de favor para moradia
df_construct['Compra p/ Morar']  = np.where((df_construct['Moradia'] == 'Aluguel') | (df_construct['Moradia'] == 'De favor'), True, False)
df_construct
```
